```python
import pandas as pd 
import numpy as np
import requests
import re
from bs4 import BeautifulSoup
```

```python
url = 'https://www.imdb.com/title/tt13751694/reviews'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
response
```

```python
print(soup)
```

```python
Review = []
rating = []
Detail_review = []
```

```python
movie_review = soup.findAll('div', attrs = {'class':'lister-item-content'})
```

```python
for store in movie_review:
    name = store.a.text
    
    rate = store.span.text

    detailed = store.find('div', class_ = 'text show-more__control').text.replace('\n','')
    
    Detail_review.append(detailed)
    
    Review.append(name)
    
    rating.append(rate)
```

```python
cleaned_rating = [r.strip('\n') for r in rating]
```

```python
cleaned_rating
```

```python
cleaned_reviews = [review.strip('\n') for review in Review]
```

```python
cleaned_reviews
```

```python
df = pd.DataFrame(columns = ["Review" ,"rating",'Detail_review'])
```

```python
df
```

```python
df['Review'] = cleaned_reviews
df['rating'] = cleaned_rating
df['Detail_review'] = Detail_review
```

```python
df.head(10)
```

```python
df['Review'] = df['Review'].str.lower()
df['Detail_review'] = df['Detail_review'].str.lower()
```

```python
df.head()
```

```python
# Get all punctuation characters
import string
exclude = set(string.punctuation)

def remove_pun(text):
    if pd.notna(text):  # Check if text is not NaN
        for char in exclude:
            text = text.replace(char, '')
    return text

# Apply the remove_pun function to the 'content' column
df['Review'] = df['Review'].apply(remove_pun)
df['Detail_review'] = df['Detail_review'].apply(remove_pun)
```

```python
df.head()
```

```python
df['rating'] = df['rating'].str.replace('/10', '')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sentiments = SentimentIntensityAnalyzer()
```

```python
sentiments = SentimentIntensityAnalyzer()
df["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in df["Review"]]
df["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in df["Review"]]
df["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in df["Review"]]
print(df.head())
```

```python
x =sum(df["Positive"])
y =sum(df["Negative"])
z =sum(df["Neutral"])

def sentiment_score(a,b,c):
    if (a>b) and (a>c):
        print("Positve ")
    elif (b>a) and (b>c):
        print("Negative ")
    else:
        print("Neutral ")
sentiment_score(x,y,z)
```
